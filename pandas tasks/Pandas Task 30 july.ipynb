```python
import pandas as pd
```

```python
performance_data=pd.read_csv("AgentPerformance.csv")
performance_data.head()
```

```python
date=pd.to_datetime(performance_data["Date"])
performance_data["date"]=date
weekly_rating=pd.DataFrame()
weekly_rating["Agent Name"]=performance_data["Agent Name"].unique()
```

```python
def week(start,end,column_name):
    mask=((performance_data["date"]>=start) & (performance_data["date"]<=end))
    weekly_data=performance_data.loc[mask]
    return (weekly_data.groupby("Agent Name")[column_name].sum()/5).reset_index()
    
```

```python
time={"start":['2022-07-01','2022-07-08','2022-07-15','2022-07-22'],"end":['2022-07-07','2022-07-14','2022-07-21','2022-07-28']}
for i in range(len(time["start"])):
    weekly_rating= pd.merge(weekly_rating,week(time["start"][i],time["end"][i],"Average Rating").rename
                            (columns={"Average Rating":f"Week_{i+1}_avg_rat"}),on="Agent Name")
```

### percentage of chat on which they have received a feedback 

```python
gp=performance_data.groupby("Agent Name")[["Total Chats","Total Feedback"]].sum()
```

```python
ag=[]
per=[]
for i in gp.index:
    if gp.loc[i,'Total Chats']==0:
        per.append(0)
        ag.append(i)
    else:
        per.append((gp.loc[i,'Total Feedback']/gp.loc[i,'Total Chats'])*100)
        ag.append(i)
```

```python
percentage=pd.DataFrame()
percentage["Agent Name"]=ag
percentage["Percentage of feedback"]=per
percentage
```
